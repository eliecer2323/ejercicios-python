```python
natalidad_pais = info.groupby(['Pais'])['Tasa de natalidad'].mean()
alta_natalidad_pais = natalidad_pais[natalidad_pais > 3]
alta_natalidad_pais
```

```python
from numpy import size
alta_natalidad_pais.plot.bar(
    title='Natalides por encima de 3'
)
plt.show()
```

```python
esperanza_vida_salud = info[['Esperanza de vida (mujeres)', 'Gasto en salud (per capita)']]
esperanza_vida_salud.plot.scatter(
    x='Esperanza de vida (mujeres)',
    y='Gasto en salud (per capita)'
)
```

```python
esperanza_vida_ano = info.groupby(['Ano'])['Esperanza de vida (mujeres)', 'Esperanza de vida (varones)'].mean().reset_index()
esperanza_vida_ano['Esperanza de vida (promedio)'] = (esperanza_vida_ano['Esperanza de vida (mujeres)'] + esperanza_vida_ano['Esperanza de vida (varones)']) / 2
esperanza_vida_ano
```

```python
esperanza_vida_ano.index = esperanza_vida_ano['Ano']
esperanza_vida_ano.plot.line(
    y='Esperanza de vida (promedio)',
    title="Esperanza de vida anual"
)
```
